```python
deliveries.groupby('delivery_status')['customer_rating_post_delivery'].mean()
```

```python
deliveries.groupby('driver_id')['manual_route_override_count'].sum().sort_values(ascending=False).head(10)
```

```python
deliveries.groupby('driver_id')['manual_route_override_count'].sum()\
.sort_values(ascending=False)\
.head(10)\
.plot(kind='bar')

plt.title('Top Drivers by Manual Route Overrides')
plt.xlabel('Driver ID')
plt.ylabel('Total Route Overrides')

plt.show()
```

```python
merged_incidents_deliveries = pd.merge(incidents, deliveries, on='delivery_id', how='left')
merged_incidents_deliveries.groupby('vehicle_id').size().sort_values(ascending=False).head(10)
```

```python
merged_incidents_deliveries.groupby('vehicle_id').size()\
.sort_values(ascending=False)\
.head(10)\
.plot(kind='bar')

plt.title('Vehicles With Most Incidents')
plt.xlabel('Vehicle ID')
plt.ylabel('Number of Incidents')

plt.show()
```

```python
delivery_complaints = pd.merge(
    deliveries,
    complaints,
    on='order_id',
    how='inner'
)

delivery_complaints.head()
```

```python
delivery_complaints.groupby('delivery_status').size()
```

```python
delivery_complaints.groupby('delivery_status').size().plot(kind='bar')

plt.title('Complaints by Delivery Status')
plt.xlabel('Delivery Status')
plt.ylabel('Number of Complaints')

plt.show()
```
